## Normal Bar chart With Color

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]


c = ['r', 'g', 'b', 'y']

plt.bar(x, y, color=c)
plt.show()
```

## Bar plot with X and Y lable with Title

```python
import matplotlib.pyplot as plt

x = ['Python', 'c', 'java', 'c++']
y = [85, 70, 60, 82]

plt.xlabel('Language', fontsize='10')
plt.ylabel('No.', fontsize='10')
plt.title('Languge demo', loc='left', fontsize='20')
plt.grid(False)

c = ['y', 'r', 'm', 'g']
plt.bar(x, y, width=0.4, color=c, align='edge',edgecolor='r',linewidth=5, linestyle='--', alpha=0.5) # color = 'r' --> all colums are red , # Aligh bydefault 'cetner' # for blur use = alpha and alpha value is only possible between 0 to 1 #

plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Python', 'c', 'java', 'c++']
y = [85, 70, 60, 82]

plt.xlabel('Language', fontsize='10')
plt.ylabel('No.', fontsize='10')
plt.title('Languge demo', loc='left', fontsize='20')
plt.grid(False)

plt.bar(x, y, width=0.4, color=c, edgecolor='r', label="laksh") # after lable always use 'legend()' to show label

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Python', 'c', 'java', 'c++']
y = [85, 70, 60, 82]
z = [40, 30, 20, 50]

plt.xlabel('Language', fontsize='10')
plt.ylabel('No.', fontsize='10')
plt.title('Languge demo', loc='left', fontsize='20')
plt.grid(False)

# Two Line in same column
plt.bar(x, y, width=0.4, color='g', label="laksh") 
plt.bar(x, z, width=0.4, color='r', label="laksh1")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['Python', 'c', 'java', 'c++']
y = [85, 70, 60, 82]
z = [40, 30, 20, 50]

width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel('Language', fontsize='10')
plt.ylabel('No.', fontsize='10')
plt.title('Languge demo', loc='left', fontsize='20')
plt.grid(False)

# if we don't want overlap but side by side then...
plt.bar(p, y, width, color='g', label="laksh") 
plt.bar(p1, z, width, color='r', label="laksh1")

plt.xticks(p + (width / 2), x, rotation=10)
plt.legend()
plt.show()
```

## Zorizontal Chart

```python
import matplotlib.pyplot as plt

x = ['Python', 'c', 'java', 'c++']
y = [85, 70, 60, 82]
z = [40, 30, 20, 50]

plt.xlabel('Language', fontsize='10')
plt.ylabel('No.', fontsize='10')
plt.title('Languge demo', loc='left', fontsize='20')
plt.grid(False)

# remove Width
plt.barh(x, y, color='g', label="laksh") 
plt.barh(x, z, color='r', label="laksh1")

plt.legend()
plt.show()
```

# 2. Scatter Plot

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
c = ['g', 'g', 'b', 'g', 'y', 'r', 'g']
sisi = [500, 40, 30, 20, 10, 30, 40]

plt.scatter(day, no, color = c, s=sisi, alpha=0.5, marker='*', edgecolors='r', linewidths=2) # s stands for size, we can do only s = 15 for all # markers are ['.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4','s', 'p', '*', 'h', 'H', '+', 'x', 'X', 'D', 'd', '|', '_', 'P']

plt.title('Scatter Plot', fontsize='20')
plt.xlabel('Day', fontsize='10')
plt.ylabel('No', fontsize='10')
plt.show()
```

# 4. Pie Plot

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ['python', 'c', 'c++', 'java']
ex = [0.2, 0.0, 0.0, 0.0]
c = ['r', 'g', 'b', 'y']

plt.pie(x, labels=y, explode=ex, colors=c, autopct='%0.1f%%', shadow=True, radius=1, labeldistance=1.1, startangle=0
        , textprops= {'fontsize': 14,'color': 'black','weight': 'bold','family': 'serif','style': 'italic'}, counterclock=True
        , wedgeprops = {'edgecolor': 'black','linewidth': 2,'linestyle': ':','width': 1.1}
) # write persentage, 0.'1' is number after point and second % is 10.0'%'<--

plt.title('Lukhi')
plt.legend(loc = 2)
plt.show()
```

# 5. Stems Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, -5, 7, -4, 3]

plt.stem(x, y, linefmt='-', markerfmt='ro', bottom=0, basefmt='g', label='python', orientation='vertical')
plt.legend()
plt.show()
```

# 6. Box Vs Whisker Plot

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70, 80]
y = [20, 30, 40, 50, 60, 70, 80, 90]

p = [x, y]

plt.boxplot(p, notch=False, vert=True, widths=0.2, labels=['python', 'c'], patch_artist='True')# vert stand for vertical # patch_artist fill color 

plt.show()
```

# 7. Stack Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 2, 5, 4]
z = [2, 3, 4, 5, 6]
a = [1, 3, 2, 4, 2]
l = ['area 1', 'area 2', 'area 3', 'area 4', 'area 5']
c = ['r', 'g', 'b']

plt.stackplot(x, y, z, a, labels=l, colors=c)

plt.title('laksh titiel')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.show()
```

# 8. Step Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y, color='r', marker = 'o', ms = 10, mfc = 'g', label='python') # ms stand for size of .(dot) 

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.grid()
plt.legend(loc = 4)
plt.show()
```
