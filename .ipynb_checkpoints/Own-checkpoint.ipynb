# pyplot Sub module of MatPlotLib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## x-pts and y-pts

```python
xpts = np.array([0, 10])
ypts = np.array([0, 125])

plt.plot(xpts, ypts)
plt.show()
```

## Chart from DataFrame

```python
df = pd.DataFrame(
    {
        "Cricket_Bat": ['SG', 'BDM', 'SS', 'GM', 'KOOKABURRA', 'SPARTAN'],
        "MRP" : [2000, 2200, 2400, 2500, 2800, 3000], 
        "Weight_Grams": [1100, 1200, 1250, 1330, 1480, 1600]
    }
)

plt.plot(df["Cricket_Bat"], df["Weight_Grams"])
# add grid line || inbuilt False
plt.grid(True)
plt.show()
```

## Add Labels to a plot

```python
plt.plot(df["MRP"], df["Weight_Grams"])
# 
plt.xlabel("Bat price(usd)")
plt.ylabel("Bat Weight(grams)")
plt.grid(True)
plt.show()
```

## Plot Title and Posstion Them

```python
plt.title('center Title')
plt.title('left Title', loc='left')
plt.title('right Title', loc='right')
```

## Add a Matplotlib legend in a graph

```python
a = np.arange(5)
b = [2, 4, 6, 8, 10]
c = [5, 6, 7, 8, 9]

ax = plt.subplot()

ax.plot(a, b, 'r--', label = 'Frequncy')
ax.plot(a, c, 'k:', label = "Periods")

# Possition of legend
ax.legend(loc = "lower right") # upper left, upper right, lower left, lower right || upper center, lower center, center left, center right
 
plt.title("Frequency of a single")

plt.show()
```

## Set Text Size and Background color of Legend

```python
a = np.arange(5)
b = [2, 4, 6, 8, 10]
c = [5, 6, 7, 8, 9]

ax = plt.subplot()

ax.plot(a, b, 'r--', label = 'Frequncy')
ax.plot(a, c, 'k:', label = "Periods")

l = ax.legend(loc = "lower right", fontsize = "xx-large") # xx-small, x-small, small, medium, large, x-large, xx-large

l.get_frame().set_facecolor('green')
 
plt.title("Frequency of a single")

plt.show()
```

## Bar graph

```python
student = np.array(['laksh', 'taksh', 'harsh', 'daksh', 'darsh', 'sarsh'])
marks = np.array([85, 59, 99, 77, 67, 95])

plt.bar(student, marks)
plt.xlabel('Names')
plt.ylabel('Marks')
plt.title("Student Result")
plt.show()
```
